```python
allfiles=sorted(glob.glob('cormfrsrM1.00.201904*.dat'))
```

```python
print(allfiles)
```

```python
df = pd.concat((pd.read_csv(f,names=head.iloc[0,:],header=None,sep=',',skiprows=9,index_col=0,engine='python') for f in allfiles))
```

```python
df.index = pd.to_datetime(df.index)
```

```python
print(df)
```

```python
df.plot(y='Dnadir(1)')
```

```python
sideavg=((df['Deast(1)']+df['Dwest(1)'])/2)
```

```python
sideavg.plot()
```

```python
sidecorr=df['Dnadir(1)']-sideavg
```

```python
sidecorr.plot()
```

```python
ax = plt.gca()
df.plot(y='Dnadir(1)',ax=ax)
sidecorr.plot(color='red',ax=ax)
```
